##Respondents who answered 'No' to BetterLife question tend to be slightly older

```python

#Look into how specific age groups answered this question and whethere answers will defer over age 50 and under age 25

over50 = df[df['Age'] >= 50]
under25 = df[df['Age'] <= 25]
```

```python
print(over50['BetterLife'].value_counts(normalize=True))
```

```python
print(under25['BetterLife'].value_counts(normalize=True))
```

```python
#Checking how large are the sample size for the age groups
print(len(over50))
print(len(under25))
```

```python
#Filtering more specific subsets
#Using pandas Boolean operators & and &~ 
# & return true if both conditions are true

#Filtering people who answered 'Yes' to the better life question & live in India
filtered_1 = df[(df['BetterLife'] == 'Yes') & (df['Country'] == 'India') ]
```

```python
filtered_1.head(5)
```

```python
#Filter for specific subsets, only people who:

    #Answered ‘Yes’ to the better life question
    #Are over age 50
    #Live in India
    #Do NOT code as a hobby
    #Contribute to open source projects at least occasionally
    
filtered_1 = df[(df['BetterLife'] == 'Yes') & (df['Age'] >= 50) & (df['Country']== 'India') &~ (df['Hobbyist'] == 'Yes') &~ (df['OpenSourcer']== 'Never') ]
filtered_1
```

Analyzing Multi-Answer Survey questions

```python
df['LanguageWorkedWith'].head()
#Respondents answer were stored in a single column, using ; as delimiter
```

```python
#Analyzing see how many developers are using python

python_bool = df['LanguageWorkedWith'].str.contains('Python')
python_bool.value_counts(normalize=True)

#42% of all developers use Python. 
```

```python
#To figure out how often each language was mentioned, split each row in series based a delimited 
# .str.split(';')
#expand=True will create a new dataframe 

lang_lists = df['LanguageWorkedWith'].str.split(';', expand=True)
lang_lists.head()
```

```python
#use df.stack() to stack dataframe which slice each column and then stack them on top of each other

lang_lists.stack().value_counts()
```

```python
#Plotting into a barchart
lang_lists.stack().value_counts().plot(kind='bar', figsize=(15,7), color='#61d199')
```

JavaScript, HTML/CSS and SQL are the top 3 prorgramming languages that respondents worked with.
